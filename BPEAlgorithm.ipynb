```python
def get_pair_counts(ids):
    """Count the frequency of all adjacent token pairs."""
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge_tokens(ids, pair, new_id):
    """Replace occurrences of a specific pair with a new token ID."""
    new_ids = []
    i = 0
    while i < len(ids):
        # If we find the target pair, merge them into the new_id
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(new_id)
            i += 2  # Skip both merged tokens
        else:
            new_ids.append(ids[i])
            i += 1
    return new_ids

def train_bpe(text, vocab_size):
    """Train BPE to build a merge dictionary and vocabulary mapping."""
    # Convert input string into a list of raw UTF-8 bytes (0-255)
    raw_bytes = list(text.encode("utf-8"))
    num_merges = vocab_size - 256
    ids = list(raw_bytes) # Copy the list to manipulate

    merges = {} # Maps (token1, token2) -> new_token_id
    
    for i in range(num_merges):
        stats = get_pair_counts(ids)
        if not stats:
            break # No pairs left to merge
            
        # Find the most frequently occurring pair
        top_pair = max(stats, key=stats.get)
        new_id = 256 + i
        
        # Merge the pair throughout the token list
        ids = merge_tokens(ids, top_pair, new_id)
        merges[top_pair] = new_id
        print(f"Merged {top_pair} into new token ID {new_id}")

    return merges
```

```python

# --- Example Demonstration ---
sample_text = "banana bandana"
desired_vocab_size = 260  # Starting with 256 basic bytes + 4 custom merges

print("Starting BPE Training...")
merges_dict = train_bpe(sample_text, desired_vocab_size)
print("\nFinal Learned Merges Dictionary:")
print(merges_dict)
```
